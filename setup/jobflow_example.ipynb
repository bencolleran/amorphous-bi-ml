```python
from jobflow import job, Flow
from jobflow.managers. local import run_locally
```

```python
@job
def add(a, b) :
    return a + b
```

```python
@job
def mult(c, d):
    return c * d
```

```python
job1 = add(1,2)
job2 = add(3,1)
job3 = mult(job1.output,job2.output)
```

```python
flow = Flow([job1,job2,job3], name="my-flow")
```

```python
flow.draw_graph(figsize=(3, 3)).show( )
```

```python
responses = run_locally(flow)
```

```python
print(responses[job2.uuid][1].output)
```
